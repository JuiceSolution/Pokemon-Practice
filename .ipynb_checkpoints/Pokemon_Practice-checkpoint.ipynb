### Creat list of colums to make Manually subsetting df easier

`df.columns`: calling all columns from the specified dataframe.
- no brackets because it's a attribute from the [dataframe object](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html).

```python
listofcol = []
for col in df.columns:
    listofcol.append(col)
```

```python
listofcol
```

### Manually Subsetting df - to information I wish to explore

`df[[columns]]`: using _**two brackets**_ to return a dataframe
- using **only one** bracket to return a [Series](https://pandas.pydata.org/docs/reference/api/pandas.Series.html).

This matters because we need to understand what kind of objects we are working with, so that we know which methods apply or not.

```python
simpledf = df[['pokedex_number','name','type1','type2','capture_rate','hp','speed','attack','sp_attack','defense','sp_defense',
               'generation','height_m','weight_kg','percentage_male','experience_growth','is_legendary', 'classfication','japanese_name']]
```

```python
simpledf
```

### Find all Pokemon That only have a single Type, Using ISNULL

`.isnull()`: returns `True` or `False` whether or not a value is null (NaN).


`df[condition]`: specifying condition(s) to query by.
- condition example: `df[col1] == value`
   

```python
singleTypepokemon = simpledf[simpledf['type2'].isnull()]
```

```python
singleTypepokemon
```

### Of the 384 'singleTypepokemon', make a subset of those from generation 1 only.

```python
STPG1 = singleTypepokemon[singleTypepokemon['generation'] == 1]
```

### Of the Generation 1 Single Type pokemon , what is the number of each single type pokemon ?

`.value_counts()`: method for Pandas.Series -> count all occurences of values. ([doc](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html))

[See also df.value_counts()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.value_counts.html): value counts for dataframe

```python
STPG1_count = STPG1['type1'].value_counts()
```

```python
STPG1_count
```

### Please create a df subset of the most common type of single type gen 1 pokemon

```python
STPG1[STPG1['type1'] == 'water']
```

## Specifying multiple conditions for querying

`df[(condition1) & (condition2) & ... (conditionN)]`: to query by multiple conditions
- note: `&` is for `and`, you can also use `|` for `or`
- example of more complex conditions:


`df[((condition1a | condition1b) & (condition2)) | (condition3)]`

```python
simpledf[(simpledf['type2'].isnull())  & (simpledf['generation'] == 1) & (simpledf['type1'] == 'water')]
```
